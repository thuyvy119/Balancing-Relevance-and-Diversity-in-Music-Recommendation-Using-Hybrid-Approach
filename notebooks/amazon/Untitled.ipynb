```python
import pandas as pd

def read_first_5_users(filename):
    """
    Read and display the first 5 users from a recommendations CSV file.
    
    Args:
        filename (str): Path to the recommendations CSV file
        
    Returns:
        pandas.DataFrame: DataFrame containing first 5 users with their recommendations
    """
    try:
        # Read the CSV file
        df = pd.read_csv(filename)
        
        print(f"Loaded recommendations from: {filename}")
        print(f"Total users in file: {len(df)}")
        print(f"Columns available: {list(df.columns)}\n")
        
        # Get first 5 users
        first_5_users = df.head(5)
        
        # Display basic info about first 5 users
        print("=" * 60)
        print("FIRST 5 USERS RECOMMENDATIONS")
        print("=" * 60)
        
        for idx, row in first_5_users.iterrows():
            print(f"\n--- User {idx+1}: {row['user_id']} ---")
            print(f"Query: {row['query']}")
            print(f"Recommendation count: {row['recommendation_count']}")
            
            # Display each recommendation for this user
            rec_count = int(row['recommendation_count'])
            for i in range(1, min(rec_count + 1, 11)):  # Show up to 10 recommendations
                title = row.get(f'rec_{i}_title', '')
                rec_id = row.get(f'rec_{i}_id', '')
                categories = row.get(f'rec_{i}_categories', '')
                
                if title:  # Only show if there's a recommendation
                    print(f"  Recommendation {i}:")
                    print(f"    Title: {title}")
                    print(f"    ID: {rec_id}")
                    print(f"    Categories: {categories}")
                    print()
            
            print("-" * 40)
        
        return first_5_users
        
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        return None
    except Exception as e:
        print(f"Error reading file: {e}")
        return None

# Usage example:
if __name__ == "__main__":
    filename = "20251120_081243_recommend_energetic_albums_with_fast_bea_recommendations.csv"
    first_5_users_df = read_first_5_users(filename)
    
    # If you want to also see the raw data structure:
    if first_5_users_df is not None:
        print("\n" + "=" * 60)
        print("RAW DATA STRUCTURE (first 5 rows):")
        print("=" * 60)
        print(first_5_users_df[['user_id', 'query', 'recommendation_count']].to_string())
```
